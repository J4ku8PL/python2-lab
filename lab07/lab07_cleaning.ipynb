## Zadanie 1b — Wynagrodzenie: 'brak' i None

```python
# 1. Sprawdź unikalne wartości w kolumnie wynagrodzenie
df['wynagrodzenie'].unique()

# 2. Zamień 'brak' na np.nan
df['wynagrodzenie'] = df['wynagrodzenie'].replace('brak', np.nan)
# 3. Skonwertuj wynagrodzenie na float używając pd.to_numeric z errors='coerce'
df['wynagrodzenie'] = pd.to_numeric(df['wynagrodzenie'], errors='coerce')
# 4. Ile NaN jest teraz w wynagrodzenie?
temp = df['wynagrodzenie'].isna().sum()
print(f'Jest {temp} Nan w kolumnie z wynagrodzeniem')
```

# Ćwiczenie 3: Pełny pipeline czyszczenia (25 min)


## Zadanie

```python
df_clean = pd.DataFrame(data) # zacznij od świeżego df
# KROK 1: Zamień 'brak' na np.nan i skonwertuj wynagrodzenie na float
# (pd.to_numeric z errors='coerce')
df_clean['wynagrodzenie'] = pd.to_numeric(df_clean['wynagrodzenie'], errors='coerce')

# KROK 2: Usuń duplikaty (drop_duplicates) i zresetuj indeks
df_clean = df_clean.drop_duplicates(subset='id_pracownika').reset_index(drop=True)
# KROK 3: Wypełnij NaN w wynagrodzeniu medianą
mediana1 = df_clean['wynagrodzenie'].median()
df_clean['wynagrodzenie'] = df_clean['wynagrodzenie'].fillna(mediana1)
# KROK 4: Oblicz średnią ocena_roczna i wypełnij NaN tą średnią
# (zaokrąglij do 2 miejsc)
avg_ocena = round(df_clean['ocena_roczna'].mean(), 2)
df_clean['ocena_roczna'] = df_clean['ocena_roczna'].fillna(avg_ocena)

# KROK 5: Wyczyść kolumnę 'imie': str.strip() + str.title()
df_clean['imie'] = df_clean['imie'].str.strip().str.title()
# KROK 6: Wyczyść kolumnę 'dzial':
# a) str.strip() + str.title()
df_clean['dzial'] = df_clean['dzial'].str.strip().str.title()
# b) str.replace('Hr', 'HR', regex=False)
df_clean['dzial'] = df_clean['dzial'].str.replace('Hr', 'HR', regex=False)
# c) str.replace('It', 'IT', regex=False)
df_clean['dzial'] = df_clean['dzial'].str.replace('It', 'IT', regex=False)
# KROK 7: Skonwertuj 'data_zatrudnienia' na datetime (errors='coerce')
df_clean['data_zatrudnienia'] = pd.to_datetime(df_clean['data_zatrudnienia'], errors='coerce')

# WERYFIKACJA — uruchom te linijki żeby sprawdzić wyniki:
print("=== WERYFIKACJA ===")
print(f"Shape: {df_clean.shape}")
print(f"\nDtypes:\n{df_clean.dtypes}")
print(f"\nNaN per kolumna:\n{df_clean.isna().sum()}")
print(f"\nUnikalne działy: {sorted(df_clean['dzial'].unique())}")
print(f"\nPierwsze imiona: {df_clean['imie'].tolist()[:8]}")
```

## Pytania biznesowe (po wyczyszczeniu)

```python
# Używając df_clean:
# Pytanie 1: Jakie jest średnie wynagrodzenie per dział?
# Wskazówka: df_clean.groupby('dzial')['wynagrodzenie'].mean().round(2)
# 1. Konwersja na liczby (coerce zamieni 'brak' i None na NaN, co pozwoli liczyć średnią)
temp4 = df_clean.groupby('dzial')['wynagrodzenie'].mean().round(2)
print(f'Średnie wynagrodzenie per dział wynosi:\n{temp4.to_string()}')

# Pytanie 2: Jaka jest średnia ocena per dział?
# Wskazówka: df_clean.groupby('dzial')['ocena_roczna'].mean().round(2)
temp5 = df_clean.groupby('dzial')['ocena_roczna'].mean().round(2)
print(f'\n Średnia ocena per dział wynosi {temp5.to_string()}')

# Pytanie 3: Kto jest najlepiej opłacany (top 3)?
# Wskazówka: df_clean.nlargest(3, 'wynagrodzenie')[['imie', 'dzial','wynagrodzenie']]
top3 = df_clean.nlargest(3, 'wynagrodzenie')[['imie', 'dzial','wynagrodzenie']]
print(f'\n Top 3 najlepiej opłacanych pracowników {top3.to_string()}')

# Pytanie 4: Ilu pracowników jest w każdym dziale?
# Wskazówka: df_clean['dzial'].value_counts()
temp6 = df_clean['dzial'].value_counts()
print(f'\n Liczba pracowników w każdym dziale {temp6.to_string()}')

```

## Rozszerzenie biznesowe 

```python
# Dodatkowe pytania dla kierownika HR:
# 1. Ile lat pracuje najdłuższy staż? (dzisiejsza data - najwcześniejsza data_zatrudnienia)
dzis = pd.Timestamp.now()
najdluzszy_staz_dni = (dzis - df_clean['data_zatrudnienia'].min()).days
najdluzszy_staz_lata = round(najdluzszy_staz_dni / 365.25, 1)
print(f'Najdłuższy staż to: {najdluzszy_staz_lata} lat')

# 2. Kto ma najwyższą ocenę w każdym dziale? (groupby + idxmax)
idx = df_clean.groupby('dzial')['ocena_roczna'].idxmax()
najlepsi_pracownicy = df_clean.loc[idx, ['dzial', 'imie', 'ocena_roczna']]
print(f'\n Pracownicy z najwyższą oceną w każdym dziale: {najlepsi_pracownicy}')
print()

# 3. Czy jest korelacja między stażem a wynagrodzeniem?
dzis = pd.Timestamp.now()
df_clean['staz_dni'] = (dzis - df_clean['data_zatrudnienia']).dt.days
korelacja = df_clean['staz_dni'].corr(df_clean['wynagrodzenie'])

print(f'\n Współczynnik korelacji wynosi: {round(korelacja, 2)}')

if korelacja > 0.7:
    print(f' Silna korelacja dodatnia (im dłuższy staż, tym wyższe zarobki).')
elif korelacja < -0.7:
    print(f' Silna korelacja ujemna (nowi pracownicy zarabiają więcej niż starzy).')
else:
    print(f' Brak silnej korelacji liniowej.')
```

## Zadanie 4b — Czyszczenie i analiza

```python
# 1. Wyczyść imie: strip() + title()
df_str['imie'] = df_str['imie'].str.strip().str.title()

# 2. Wyczyść dzial: strip() + title() + replace
df_str['dzial'] = df_str['dzial'].str.strip().str.title()
df_str['dzial'] = df_str['dzial'].str.replace('Hr', 'HR', regex=False)
df_str['dzial'] = df_str['dzial'].str.replace('It', 'IT', regex=False)

# 3. Ile pracowników ma imię zaczynające się na 'A'?
temp7 = df_str[df_str['imie'].str.startswith('A')]
print(f"\n Liczba osób na literę 'A': {temp7.to_string()}")
# 4. Utwórz kolumnę 'dlugosc_imienia' z długością imienia
df_str['dlugosc_imienia'] = df_str['imie'].str.len()
temp8 = df_str['dlugosc_imienia'].describe()
print(f'\n {temp8.to_string()}')
# 5. Kto ma najdłuższe imię? (nlargest lub sort_values)

temp9 = df_str.nlargest(1, 'dlugosc_imienia')[['imie', 'dlugosc_imienia']]
print(f"\nOsoba z najdłuższym imieniem: {temp9}")
```

## Zadanie 4c — str.contains z wzorcem

```python
# str.contains() może używać wyrażeń regularnych!
# 1. Znajdź pracowników z działów IT lub HR (jeden zawiera drugi wzorzec)

temp10 = df_str[df_str['dzial'].str.contains('IT|HR')]
print(f'\n Pracownicy IT oraz HR: \n {temp10}')

# 2. Znajdź imiona zawierające literę 'a' (małą lub dużą)
temp11 = df_str[df_str['imie'].str.contains('a', case=False)]
print(f"\n Liczba imion zawierających literę 'a': {len(temp11)}")
# 3. Ile pracowników ma imię kończące się na 'a'?
# Wskazówka: str.endswith('a') lub str.contains('a$', regex=True)
temp12 = df_str['imie'].str.contains('a$', regex=True).sum()
print(f"\nLiczba pracowników, których imię kończy się na 'a': {temp12}")
```
